```python
# Import RiveScript
import rivescript as rv
import requests as rq
import Gruppen.Jokebot.weather as weather
import Gruppen.Jokebot.jokes as jokes
```

```python
# Print all Methods/Variables of an object (for debug purposes)
def get_fields(obj):
    fields = [method_name for method_name in dir(obj) if callable(getattr(obj, method_name))]
    print(fields)
```

```python
# rv_ variables are the actual rivescript bots
rv_nott = ""
rv_rick = ""
rv_joke = ""
rv_elon = ""

# these functions reset a bot completely
def reset_bot_nott():
    global rv_nott
    rv_nott = rv.RiveScript()
    rv_nott.load_file("./Gruppen/Nott.rive")
    rv_nott.sort_replies()
    
def reset_bot_rick():
    global rv_rick
    rv_rick = rv.RiveScript()
    rv_rick.load_file("./Gruppen/Rick.rive")
    rv_rick.sort_replies()
        
def reset_bot_joke():
    global rv_joke
    rv_joke = rv.RiveScript()
    rv_joke.load_file("./Gruppen/Jokebot/brain/main.rive")
    rv_joke.sort_replies()
    
    rv_joke.set_subroutine("setup_joke", jokes.setup_joke)
    rv_joke.set_subroutine("random_joke", lambda rs, s: jokes.random_joke())
    rv_joke.set_subroutine("random_chuck_norris", lambda rs, s: jokes.random_chucknorris_joke())
    rv_joke.set_subroutine("its_today", lambda rs, s: weather.its_today(s[0]))
    rv_joke.set_subroutine("get_temperature", lambda rs, s: weather.get_temperature())
    rv_joke.set_subroutine("todays_weather", lambda rs, s: weather.todays_weather())

def reset_bot_elon():
    global rv_elon
    rv_elon = rv.RiveScript()
    rv_elon.load_file("./Gruppen/Elon.rive")
    rv_elon.sort_replies()
```

```python
# start all bots
reset_bot_nott()
reset_bot_rick()
reset_bot_joke()
reset_bot_elon()
```

```python
# oh my god this code is a mess, so sorry
import gradio as gr


chat_history = []
current_bot = "Nott"

# change the current bot to a different one
def set_bot_nott():
    global current_bot
    global chat_history
    
    reset_bot_nott()
    current_bot = "Nott"
    chat_history = []
    
    print("Set Bot to " + current_bot)
    
    return chat_history
    
def set_bot_rick():
    global current_bot
    global chat_history
    
    reset_bot_rick()
    current_bot = "Rick"
    chat_history = []
    
    print("Set Bot to " + current_bot)
    
    return chat_history
    
def set_bot_joke():
    global current_bot
    global chat_history
    
    reset_bot_joke()
    current_bot = "Joke"
    chat_history = []
    
    print("Set Bot to " + current_bot)
    
    return chat_history
    
def set_bot_elon():
    global current_bot
    global chat_history
    
    reset_bot_elon()
    current_bot = "Elon"
    chat_history = []
    
    print("Set Bot to " + current_bot)
    
    return chat_history

# get reply and store message and reply in chat_history
def rive_reply(message):
    
    # print("Getting Reply from Bot " + current_bot) 
    
    # get response from current bot
    if current_bot == "Nott":
        response = rv_nott.reply("Tony", message)
        
    elif current_bot == "Rick":
        response = rv_rick.reply("Tony", message)
        
    elif current_bot == "Joke":
        response = rv_joke.reply("Tony", message)
    
    elif current_bot == "Elon":
        response = rv_elon.reply("Tony", message)
        
    else:
        response = "ERROR! No Bot with name " + current_bot + " found!"
        
    # append message to array
    chat_history.append((message, response))
    
    return [chat_history, ""]

# build UI
with gr.Blocks() as demo:
    
    with gr.Box():
        
        with gr.Column():
            
            chat_window = gr.Chatbot()
            chat_window.color_map = (('lime', 'gray'))

            input_field = gr.Textbox(label="Send a Message")
            
            button_send = gr.Button(value="Send",variant="primary")
            
            # buttons for changing the bot
            with gr.Row():
                button_bot1 = gr.Button(value="Nott")
                button_bot2 = gr.Button(value="B0TR1CK")
                button_bot3 = gr.Button(value="Friendly Chatbot")
                button_bot4 = gr.Button(value="Ask Elon")
                
                button_bot1.click(fn=set_bot_nott, inputs=None, outputs=chat_window)
                button_bot2.click(fn=set_bot_rick, inputs=None, outputs=chat_window)
                button_bot3.click(fn=set_bot_joke, inputs=None, outputs=chat_window)
                button_bot4.click(fn=set_bot_elon, inputs=None, outputs=chat_window)

            # submit value via click
            button_send.click(
                fn=rive_reply,
                inputs=input_field,
                outputs=[chat_window, input_field],
            )
            
            # submit value via enter
            input_field.submit(
                fn=rive_reply,
                inputs=input_field,
                outputs=[chat_window, input_field],            
            )

demo = demo.launch(inline=False, inbrowser=True)
```
